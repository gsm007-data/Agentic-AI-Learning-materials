Automated Machine Learning (AutoML) refers to a set of tools and techniques that automate the end-to-end process of building machine learning models. Instead of manually handling each step—like data preprocessing, feature engineering, model selection, and hyperparameter tuning—AutoML systems perform these tasks automatically with minimal human intervention.

The primary goal of AutoML is to make machine learning faster, more accessible, and less error-prone, especially for teams that want reliable models without deep ML expertise.

- Ref - https://www.datacamp.com/blog/best-automl-frameworks

## autogluon

AutoGluon is an open-source Python AutoML framework developed by AWS AI that automates machine learning tasks with a focus on high accuracy, minimal code, and support for tabular, text, and image data. 

It builds a diverse set of models and uses automated model selection, hyperparameter tuning, and ensemble learning to deliver strong predictive performance across different data types.

Reference - https://github.com/autogluon/autogluon

```python
from autogluon.tabular import TabularDataset, TabularPredictor
import matplotlib.pyplot as plt

label = "class"
train_data = TabularDataset("train_data.csv")

predictor = TabularPredictor(label=label).fit(train_data)

test_data = TabularDataset("test_data.csv")
predictions = predictor.predict(test_data.drop(columns=[label]))
plt.show()
```
