```python
# Install spaCy and download a model (if not already done)
# pip install spacy
# python -m spacy download en_core_web_md

import spacy

# Load medium English model (has word vectors)
nlp = spacy.load("en_core_web_md")

# Define some gendered and profession words
words = ["man", "woman", "programmer", "nurse", "engineer", "teacher", "cook"]

# Helper function to compute cosine similarity
def similarity(w1, w2):
    return nlp(w1).vector.dot(nlp(w2).vector) / (
        nlp(w1).vector_norm * nlp(w2).vector_norm
    )

# Compare how close professions are to gender terms
for profession in ["programmer", "nurse", "engineer", "teacher", "cook"]:
    sim_to_man = similarity(profession, "man")
    sim_to_woman = similarity(profession, "woman")
    print(f"{profession.capitalize():<12} → man: {sim_to_man:.3f}, woman: {sim_to_woman:.3f}")
```
